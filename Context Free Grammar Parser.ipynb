```python
from pprint import pprint
import time

# END OF SECTION

def read_cfg(input_file):
    with open(input_file, 'r') as f:
        read_data = f.readlines() # Turn the input into a list
        
    global input_as_list # Make the input data available to the rest of hte application
    input_as_list = [i.strip('\n') for i in read_data if len(i) > 1] # Remove new-line characters and empty lines from the input

    
class reader(object):    
    def __init__(self, strategy=None):
        self.action = strategy # Default strategy is set to 'None' in initialization of reader() object
            
    def set_read_type(self, read_type):
        self.action = read_type # Define which object will handle the reading of the line
            
    def read(self, line_in):
        self.action.read(self, line_in) # Use the read method of the currently defined strategy
    
global variables
global terminals
global rules
global start

class read_tuples: # Interface for the read method
    def read(self):
        pass

class read_variables(read_tuples): # read in variables from a comma delimited line
    @staticmethod
    def read(self, line_in):
        variables = line_in.split(",")
        print("reading variables: ", variables)
        print(variables)
    
class read_terminals(read_tuples): # read in terminals from a comma delimited line
    @staticmethod
    def read(self, line_in):
        terminals = line_in.split(",")
        print("reading terminals: ", terminals)
        print(terminals)
    
class read_rules(read_tuples): # read a single rule and append it to the list of rules
    @staticmethod
    def read(self, line_in):
        rules.append(line_in)
        print("reading rule: ", line_in)
    
class read_start(read_tuples): # read the start variable
    @staticmethod
    def read(self, line_in):
        start = line_in
        print("reading start variable: ", start)
    
    
def read_input(): 
    r = reader()
    for i in input_as_list:
        if i == "/VARIABLES":
            print(i)
            r.set_read_type(read_variables())
        elif i == "/TERMINALS":
            print(i)
            r.set_read_type(read_terminals())
        elif i == "/RULES":
            print(i)
            r.set_read_type(read_rules())
        elif i == "/START":
            print(i)
            r.set_read_type(read_start())
        else:
            r.read(i)
    
    
read_cfg("G0.txt")
read_input()
print(variables)
print(terminals)
print(rules)
print(start)

cfg1 = CFG(variables, terminals, rules, start)
print(cfg1.variables)
print(cfg1.terminals)
print(cfg1.rules)
print(cfg1.start)
```
